H4: Write a program to take two string numbers and convert them to integers
using explicit type conversion.


```python
num1 = input("Enter the num1: ")
num2 = input("Enter the num2: ")

print("Before conversion:")
print("num1 =", num1, "Type:", type(num1))
print("num2 =", num2, "Type:", type(num2))

num1 = int(num1)
num2 = int(num2)

print("\nAfter conversion:")
print("num1 =", num1, "Type:", type(num1))
print("num2 =", num2, "Type:", type(num2))

print("\nSum:", num1 + num2)
```

H5: Write a program to calculate and print the average of the three
numbers. Round the result to two decimal places.


```python
a = float(input("Enter first number: "))
b = float(input("Enter second number: "))
c = float(input("Enter third number: "))

print("\nTypes of inputs:")
print("a Type:", type(a))
print("b Type:", type(b))
print("c Type:", type(c))

avg = (a + b + c) / 3

print("\nAverage:", round(avg, 2))
print("Average Type:", type(avg))
```

H6: Create a simple calculator that performs basic arithmetic operations
(addition, subtraction, multiplication, division) based on user input.


```python
a = float(input("Enter the number value for a: "))
b = float(input("Enter the number value for b: "))
op = input("Enter operation (add, sub, multi, div): ")

if op == "add":
    print("Result:", a + b)
elif op == "sub":
    print("Result:", a - b)
elif op == "multi":
    print("Result:", a * b)
elif op == "div":
    if b != 0:
        print("Result:", a / b)
    else:
        print("Cannot divide by zero")
else:
    print("Please enter the approriate operation")
```

H8: Create a program that converts temperatures between Fahrenheit and
Celsius based on user input.

```python
choice = input("Convert \n1.Celsius to Fahrenheit  \n2.Fahrenheit to Celsius: ")

if choice == "1":
    c = float(input("Enter Celsius: "))
    f = (c * 9/5) + 32
    print("Fahrenheit:", f)
elif choice == "2":
    f = float(input("Enter Fahrenheit: "))
    c = (f - 32) * 5/9
    print("Celsius:", c)
else:
    print("Invalid choice")
```

H9: Write a program that asks the user to enter three numbers and
determines which one is the largest.


```python
a = float(input("Enter the num value of a: "))
b = float(input("Enter the num value of b: "))
c = float(input("Enter the num value of c: "))

if a >= b and a >= c:
    print("Largest:", a)
elif b >= a and b >= c:
    print("Largest:", b)
else:
    print("Largest:", c)
```

H10: Write a program that prompts the user to input a number N and prints
all even numbers between 1 and N using a for loop.


```python
n = int(input("Enter the value of N: "))

print("The even numbers from 1 to", n, "are:")

for i in range(1, n+1):
    if i % 2 == 0:
        print(i)
```
