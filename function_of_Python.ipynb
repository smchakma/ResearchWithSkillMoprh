#H1. Write a function to calculate the sum of numbers from 1 to N.

```python
def sum_to_n(n):
  sum =0
  for i in range(1, n+1):
    sum += i
  return sum

print(sum_to_n(50))
```

#H2. Write a function that returns the square of a number.

```python
def square(n):
  return n*n

print(square(5))
```
